```python
beef_and_pork_arr = df3[['beef', 'pork']].to_numpy()
beef_and_pork_arr
```

```python
# Display beef_and_pork_arr as Line Plot
sns.axes_style("darkgrid")
sns.set_theme(style="darkgrid")
plot = sns.lineplot(data=beef_and_pork_arr, scalex=True, scaley=True, dashes=False, markers=True, legend="full")
plot.set(xlabel="Year", ylabel="Quantity")
plt.xticks(ticks=list(range(1, 51)),labels=df3['code'].to_list(), rotation=60, fontsize=6)
plt.grid(True)
plt.show()
```
